```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Simulate data
np.random.seed(42)

industries = ['Retail', 'Manufacturing', 'Healthcare', 'Energy', 'Agriculture']
products = ['Air Conditioning', 'Energy Systems', 'Water Management', 'Cooling Systems', 'Heating Systems']
regions = ['North America', 'Europe', 'Asia', 'Africa', 'Oceania']

data = {
    'Industry': np.random.choice(industries, 100),
    'Product': np.random.choice(products, 100),
    'Region': np.random.choice(regions, 100),
    'Profit': np.random.randint(50000, 200000, size=100),
    'Cooling Degree Days': np.random.randint(100, 300, size=100)  # Simulate Cooling Degree Days (CDD)
}

df = pd.DataFrame(data)
df.head()  # Display the first few rows of the dataset
```

```python
# Pie chart for profit distribution by industry
industry_profit = df.groupby('Industry')['Profit'].sum()

plt.figure(figsize=(7, 7))
plt.pie(industry_profit, labels=industry_profit.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Profit Distribution by Industry')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Scatter plot of profit vs Cooling Degree Days by region
plt.figure(figsize=(10, 6))
for region in df['Region'].unique():
    region_data = df[df['Region'] == region]
    plt.scatter(region_data['Cooling Degree Days'], region_data['Profit'], label=region, alpha=0.6)

plt.title('Profit vs Cooling Degree Days by Region')
plt.xlabel('Cooling Degree Days (CDD)')
plt.ylabel('Profit')
plt.legend(title='Region')
plt.show()
```

```python
# Bar chart showing profit distribution by region
region_profit = df.groupby('Region')['Profit'].sum()

plt.figure(figsize=(10, 6))
region_profit.plot(kind='bar', color=plt.cm.Paired.colors)
plt.title('Profit Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.show()
```
